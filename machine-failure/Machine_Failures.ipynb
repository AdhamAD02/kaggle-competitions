```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler , Normalizer , MinMaxScaler , OneHotEncoder
from sklearn.model_selection import train_test_split

from xgboost import XGBClassifier
```

```python
train_data = pd.read_csv('/home/adham/kaggle competitions/machine-failure/train.csv')
test_data = pd.read_csv('/home/adham/kaggle competitions/machine-failure/test.csv')
train_data.head()
```

```python
test_data.head()
```

```python
y = train_data['Machine failure']
test_id = pd.DataFrame(test_data['id'])
train_data.drop(columns=['id' , 'Product ID' , 'Machine failure'] , inplace = True)
test_data.drop(columns=['id' , 'Product ID'] , inplace = True)
Xpca_test = test_data.iloc[ : , -5:]
Xpca = train_data.iloc[: , -5:]

```

```python
pca = PCA(n_components=2)  # You can adjust the number of components
X_pca = pca.fit_transform(Xpca)
X_pca_test = pca.fit_transform(Xpca_test)
X_pca
```

```python
train_data[['TWF' , 'HDF']] = X_pca
train_data.drop(columns = ['PWF' ,	'OSF' ,	'RNF'] , inplace = True)

test_data[['TWF' , 'HDF']] = X_pca_test
test_data.drop(columns = ['PWF' ,	'OSF' ,	'RNF'] , inplace = True)

test_data
```

```python
numeric = ['Air temperature [K]' ,	'Process temperature [K]' , 	'Rotational speed [rpm]',	'Torque [Nm]' , 	'Tool wear [min]']
cat = ['Type']
passThrough = ['TWF' , 'HDF']

pipe_num = Pipeline(steps = [
    ('scale' , StandardScaler())
])

pipe_cat = Pipeline(steps = [
    ('cat' , OneHotEncoder(handle_unknown='ignore'))
])

columnTransformer = ColumnTransformer(transformers = [
    ('num' , pipe_num , numeric) ,
    ('catg' , pipe_cat , cat),
    ('passthrough' , 'passthrough' , passThrough )
])

columnTransformer.fit(train_data)
X_train = columnTransformer.transform(train_data)
columnTransformer.fit(test_data)
X_test = columnTransformer.transform(test_data)
```

```python
X_train , X_testing , y_train , y_testing = train_test_split(X_train , y , test_size = 0.2 , random_state = 42)
```

```python
xgb = XGBClassifier()
xgb.fit(X_train, y_train)
sc = xgb.score(X_testing,y_testing)
sc
```

```python
y_pred = xgb.predict(X_test)
test_id['Machine failure'] = pd.DataFrame(y_pred)
test_id.to_csv('XGB.csv')
```
